```python
# Big scale LLs

ILL = '/zfs/nrodd/NPTF-ID-Catalog/SimpleScan/data/FloatPS_indiv_floatDM_dm3/LL_inten_o0_mc'

ncols=5
nrows=8
nmc = 5

fig, axes2d = plt.subplots(nrows=nrows, ncols=ncols,sharex=True, sharey=True,figsize=(30/1.5,20/1.5))

for i, row in tqdm(enumerate(axes2d)):
    for j, cell in enumerate(row):
        iebin = 5*i+j
        LL2 = np.zeros((nmc, 200))
        inten = np.load(ILL+'0.npz')['intens'][iebin]
        for imc in range(nmc):
            LL2[imc] = 2*np.load(ILL+str(imc)+'.npz')['LL'][iebin]
            LL2[imc]-= LL2[imc][0]
        
        cell.fill_between(inten,np.percentile(LL2,2.5,axis=0),np.percentile(LL2,97.5,axis=0),color='firebrick',alpha=0.1)
        cell.fill_between(inten,np.percentile(LL2,16,axis=0),np.percentile(LL2,84,axis=0),color='firebrick',alpha=0.15)
        cell.plot(inten,np.percentile(LL2,50,axis=0),color='firebrick')
        cell.set_ylabel("TS")
        cell.set_xlabel("Intensity")
        cell.set_xscale("log")
        #cell.set_xlim(1e-11,1e-8)
        cell.text(1.5e-17,150, "Ebin " + str(iebin))
        cell.set_ylim(-5,1000)
```

```python
mc_dir = '/zfs/nrodd/NPTF-ID-Catalog/SimpleScan/data/FloatPS_indiv_floatDM_dm3/LL2_TSmx_lim_b_o'
nmc=5

# Catalog used for the analysis (e.g. 2MASS or DarkSky)
catalog_file = '/zfs/nrodd/NPTF-ID-Catalog/data/DarkSky_ALL_200,200,200_v3.csv'

# Cuts
halos_to_keep = 100 # total number of halos to keep
bcut = 20 # latitude cut
cut_0p5 = False # cut halos within 0.5 degrees of a 3FGL source
nonoverlap = False # whether to insist on non-overlapping regions
nonoverlapradius = 2 # radius of non-overlapping region

# Default TS xsec parameters
TS100 = 5
TS1000 = 9
TSabove = 16
xsecslim = 10

# Define the xsec and mass array, these should be the same as used during the runs
xsecs = np.logspace(-33,-18,301)
marr = np.array([1.00000000e+01,1.50000000e+01,2.00000000e+01,2.50000000e+01,3.00000000e+01,
                 4.00000000e+01,5.00000000e+01,6.00000000e+01,7.00000000e+01,8.00000000e+01,
                 9.00000000e+01,1.00000000e+02,1.10000000e+02,1.20000000e+02,1.30000000e+02,
                 1.40000000e+02,1.50000000e+02,1.60000000e+02,1.80000000e+02,2.00000000e+02,
                 2.20000000e+02,2.40000000e+02,2.60000000e+02,2.80000000e+02,3.00000000e+02,
                 3.30000000e+02,3.60000000e+02,4.00000000e+02,4.50000000e+02,5.00000000e+02,
                 5.50000000e+02,6.00000000e+02,6.50000000e+02,7.00000000e+02,7.50000000e+02,
                 8.00000000e+02,9.00000000e+02,1.00000000e+03,1.10000000e+03,1.20000000e+03,
                 1.30000000e+03,1.50000000e+03,1.70000000e+03,2.00000000e+03,2.50000000e+03,
                 3.00000000e+03,4.00000000e+03,5.00000000e+03,6.00000000e+03,7.00000000e+03,
                 8.00000000e+03,9.00000000e+03,1.00000000e+04])

# Load the catalog and extract the b values for the latitude cut, l for the ROI cut (if specified),
# apply 0p5 cut if required
catalog = pd.read_csv(catalog_file)[:halos_to_keep]
b_array = catalog.b.values
ell_array = catalog.l.values
near_3FGL = np.array([ len(catalog[u'3FGL 0.5'].values[i]) for i in range(len(b_array))])

if cut_0p5:
    good_vals = np.where((np.abs(b_array) > bcut) & (near_3FGL == 2))[0]
else:
    good_vals = np.where(np.abs(b_array) > bcut)[0]

good_halos = len(good_vals)
```

```python
MC_limit_arr = np.zeros((len(marr),nmc))

for imc in range(nmc):
    LL2_cumulative = np.zeros((len(marr),len(xsecs)))
    for iobj in tqdm(good_vals):

        # Load halo max TS, mloc and xsec values and see if passes cut
        pp_file = np.load(mc_dir + str(iobj) + '_mc' + str(imc) + '.npz')

        # Add this objects LL at the relevant mass to the cumulative LL 
        LL2_cumulative += pp_file['LL2']

        # If have enough halos stop, otherwise continue
        if halos_kept == halos_to_keep: break

    # Now calculate the limit at each mass point for this MC
    for im in range(len(marr)):
        TS_xsec_ary = LL2_cumulative[im] - LL2_cumulative[im][0]

        max_loc = np.argmax(TS_xsec_ary)
        max_TS = TS_xsec_ary[max_loc]
        for xi in range(max_loc,len(xsecs)):
            val = TS_xsec_ary[xi] - max_TS
            if val < -2.71:
                # Save log10 of the limit
                scale = (TS_xsec_ary[xi-1]-max_TS+2.71)/(TS_xsec_ary[xi-1]-TS_xsec_ary[xi])
                MC_limit_arr[im,imc] = np.log10(xsecs[xi-1])+scale*(np.log10(xsecs[xi])-np.log10(xsecs[xi-1]))
                break

################
# MC Quartiles #
################

MC_lim_p2p5 = np.percentile(MC_limit_arr, 2.5, axis = 1) 
MC_lim_p16 = np.percentile(MC_limit_arr, 16, axis = 1) 
MC_lim_p84 = np.percentile(MC_limit_arr, 84, axis = 1) 
MC_lim_p97p5 = np.percentile(MC_limit_arr, 97.5, axis = 1) 
```
