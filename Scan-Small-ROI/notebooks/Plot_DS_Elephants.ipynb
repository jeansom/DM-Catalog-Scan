```python
import sys
sys.path.append("../")

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tqdm import *
from plot import LimitPlot

%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
plots_dir = "/tigress/nrodd/DM-Catalog-Scan/Scan-Small-ROI/plots_darksky/"
```

# Initialize

```python
plot_mc_sl0 = LimitPlot(data_dir='/tigress/nrodd/DM-Catalog-Scan/Scan-Small-ROI/data/DarkSky_sid_loc0/',
                        elephant=True, 
                        nmc=100, 
                        halos_ran=1000, 
                        halos_to_keep=1000,
                        data_type="mc",
                        bcut=20,
                        nonoverlap=True,
                        nonoverlapradius=2.,
                        cut_0p5=False,
                        xsecslim=10,
                        TS100=4,
                        TS1000=9,
                        elephantm = [0,11,52])

plot_mc_sl1 = LimitPlot(data_dir='/tigress/nrodd/DM-Catalog-Scan/Scan-Small-ROI/data/DarkSky_sid_loc1/',
                        elephant=True, 
                        nmc=100, 
                        halos_ran=1000, 
                        halos_to_keep=1000,
                        data_type="mc",
                        bcut=20,
                        nonoverlap=True,
                        nonoverlapradius=2.,
                        cut_0p5=False,
                        xsecslim=10,
                        TS100=4,
                        TS1000=9,
                        elephantm = [0,11,52])

plot_mc_sl2 = LimitPlot(data_dir='/tigress/nrodd/DM-Catalog-Scan/Scan-Small-ROI/data/DarkSky_sid_loc2/',
                        elephant=True, 
                        nmc=100, 
                        halos_ran=1000, 
                        halos_to_keep=1000,
                        data_type="mc",
                        bcut=20,
                        nonoverlap=True,
                        nonoverlapradius=2.,
                        cut_0p5=False,
                        xsecslim=10,
                        TS100=4,
                        TS1000=9,
                        elephantm = [0,11,52])

plot_mc_sl3 = LimitPlot(data_dir='/tigress/nrodd/DM-Catalog-Scan/Scan-Small-ROI/data/DarkSky_sid_loc3/',
                        elephant=True, 
                        nmc=100, 
                        halos_ran=1000, 
                        halos_to_keep=1000,
                        data_type="mc",bcut=20,
                        nonoverlap=True,
                        nonoverlapradius=2.,
                        cut_0p5=False,
                        xsecslim=10,
                        TS100=4,
                        TS1000=9,
                        elephantm = [0,11,52])
```

# Plot

```python
# Loc 0
mc_ary = plot_mc_sl0.return_limits()

for midx in [0,11,52]:
    plt.plot(np.arange(len(mc_ary[0][:,midx,:])) + 1,np.percentile(mc_ary[0][:,midx,:],[50],axis=1)[0], color='cornflowerblue', ls='--')

    plt.fill_between(np.arange(len(mc_ary[0][:,midx,:])) + 1,np.percentile(mc_ary[0][:,midx,:],[16],axis=1)[0],np.percentile(mc_ary[0][:,midx,:],[84],axis=1)[0], alpha=0.1, color='cornflowerblue')
    plt.fill_between(np.arange(len(mc_ary[0][:,midx,:])) + 1,np.percentile(mc_ary[0][:,midx,:],[2.5],axis=1)[0],np.percentile(mc_ary[0][:,midx,:],[97.5],axis=1)[0], alpha=0.2, color='cornflowerblue')

    plt.title("$M_\chi$ = " + str(int(plot_mc_sl0.marr[midx])) + " GeV")
    
    plt.xlabel("$N_h$")
    plt.ylabel("$\langle \sigma v \\rangle$ [cm$^3$s$^{-1}$]")
    
    plt.xscale("log")
    plt.show()

np.save("mc_ary_loc0.npy",mc_ary)
```

```python
save_dir = '/tigress/bsafdi/github/NPTF-working/NPTF-ID-Catalog/runs_sid/notebooks/DataArrays/'

oFig1 = plt.figure(1,figsize=(15,10))

for iloc in range(4):
    mc_ary = np.load("mc_ary_loc"+str(iloc)+".npy")
    for imass, midx in enumerate([0,11,52]):
        oFig1.add_subplot(4,3,3*iloc+imass+1)
        
        plt.plot(np.arange(len(mc_ary[0][:,midx,:])) + 1,10**np.percentile(mc_ary[0][:,midx,:],[50],axis=1)[0], color='tomato', ls='--')

        plt.fill_between(np.arange(len(mc_ary[0][:,midx,:])) + 1,10**np.percentile(mc_ary[0][:,midx,:],[16],axis=1)[0],10**np.percentile(mc_ary[0][:,midx,:],[84],axis=1)[0], alpha=0.1, color='tomato')
        plt.fill_between(np.arange(len(mc_ary[0][:,midx,:])) + 1,10**np.percentile(mc_ary[0][:,midx,:],[2.5],axis=1)[0],10**np.percentile(mc_ary[0][:,midx,:],[97.5],axis=1)[0], alpha=0.2, color='tomato')

        if iloc == 0:
            if imass == 2: 
                plt.title("\\textbf{\emph{m}}$\mathbf{_\chi}$ \\textbf{= 10 TeV}",fontsize=25)
            else:
                plt.title("\\textbf{\emph{m}}$\mathbf{_\chi}$ \\textbf{= " + str(int(plot_mc_sl3.marr[midx])) + " GeV}",fontsize=25) 
        if iloc == 3:
            plt.xlabel("$N_h$")
            
        if imass == 0:
            plt.ylabel("\\textbf{Location "+str(iloc +1) + "}\n$\langle \sigma v \\rangle$ [cm$^3$s$^{-1}$]",fontsize=25)

        plt.xscale("log")
        plt.yscale("log")

        plt.xticks(fontsize=23);
        plt.yticks(fontsize=23);
        
        plt.xlim(1, 400)
        
        if imass == 0:
            plt.ylim(1e-26,1e-24)
        elif imass ==1:
            plt.ylim(1e-25,1e-23)
        elif imass ==2:
            plt.ylim(1e-22,1e-20)
            

        plt.tight_layout()

plt.savefig(plots_dir + "darksky_elephants.pdf")

```
