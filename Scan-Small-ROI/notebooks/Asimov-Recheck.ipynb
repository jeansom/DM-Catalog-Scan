```python
import sys
sys.path.append("../")

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tqdm import *
from plot import LimitPlot

%matplotlib inline
%load_ext autoreload
%autoreload 2
```

# Location 0

## Load files

```python
plot_mc = LimitPlot(data_dir='/tigress/nrodd/DM-Catalog-Scan/Scan-Small-ROI/data/DarkSky_ben_together_nodm_loc0/',
                    elephant=True,
                    nmc=80, 
                    halos_ran=100, 
                    halos_to_keep=100,
                    TS100=1e5,
                    TS1000=1e5,
                    TSabove=1e5,
                    xsecslim=1e5,
                    data_type="mc",
                    elephantm = [0,11,52])

plot_Asimov = LimitPlot(data_dir='/tigress/nrodd/DM-Catalog-Scan/Scan-Small-ROI/data/Asimov_together_nodm_loc0/',
                        elephant=True,
                        halos_ran=100, 
                        halos_to_keep=100,
                        TS100=1e5,
                        TS1000=1e5,
                        TSabove=1e5,
                        xsecslim=1e5,
                        data_type="Asimov",
                        elephantm = [0,11,52])

# Check files all there
imc_missing, iobj_missing = plot_mc.files_exist()
imc_missing, iobj_missing = plot_Asimov.files_exist()

# Calculate Limits
mc_ary = plot_mc.return_limits()
Asimov_ary = plot_Asimov.return_limits()
```

## Plot

```python
# Setup 1/J and 1/J^2 arrays
catalog = pd.read_csv('/tigress/nrodd/DM-Catalog-Scan/DataFiles/Catalogs/DarkSky_ALL_200,200,200_v3.csv')
good_vals = plot_mc.good_vals
J_values_load = np.array(10**(catalog['mulog10J_true'].values[good_vals]))
J_values = J_values_load/J_values_load[0]

for midx in [0,11,52]:
    # Calculate high and low stat Asimov scaling
    base_lim = Asimov_ary[0][:,midx,:][0,0]
    lim_highstat = np.zeros(len(J_values))
    lim_lowstat = np.zeros(len(J_values))

    for i in range(len(J_values)):
        lim_highstat[i] = np.log10((10**base_lim)/np.sum(J_values[:i]))
        lim_lowstat[i] = np.log10((10**base_lim)/np.sqrt(np.sum(J_values[:i]**2)))
    
    plt.plot(np.arange(len(mc_ary[0][:,midx,:])) + 1,np.percentile(mc_ary[0][:,midx,:],[50],axis=1)[0], color='cornflowerblue', ls='--',label='MC')
    plt.plot(np.arange(len(mc_ary[0][:,midx,:])) + 1,Asimov_ary[0][:,midx,:][:,0], color='black', ls='-',label='Asimov')
    plt.plot(np.arange(len(mc_ary[0][:,midx,:])),lim_highstat, color='black', ls='--',label='1/J')
    plt.plot(np.arange(len(mc_ary[0][:,midx,:])),lim_lowstat, color='black', ls='dotted',label='1/sqrt(J**2)')

    plt.fill_between(np.arange(len(mc_ary[0][:,midx,:])) + 1,np.percentile(mc_ary[0][:,midx,:],[16],axis=1)[0],np.percentile(mc_ary[0][:,midx,:],[84],axis=1)[0], alpha=0.1, color='cornflowerblue')
    plt.fill_between(np.arange(len(mc_ary[0][:,midx,:])) + 1,np.percentile(mc_ary[0][:,midx,:],[2.5],axis=1)[0],np.percentile(mc_ary[0][:,midx,:],[97.5],axis=1)[0], alpha=0.2, color='cornflowerblue')

    plt.title("$M_\chi$ = " + str(int(plot_mc.marr[midx])) + " GeV")
    
    plt.xlabel("$N_h$")
    plt.ylabel("$\langle \sigma v \\rangle$ [cm$^3$s$^{-1}$]")
    
    if midx == 0:
        plt.ylim(-26,-24)
    elif midx == 11:
        plt.ylim(-25,-23)
    else:
        plt.ylim(-22.5,-20.5)
    
    plt.xscale("log")
    plt.show()
```
